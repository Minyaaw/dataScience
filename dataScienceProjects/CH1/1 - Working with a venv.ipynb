This notebook helps you install extra libraries in the virtual environment that you created when you followed the powerpoint called '1bis Setting up your working environment'.

You can install all libraries one by one. Let's do that with a library used for plotting, called matplotlib:

```python
!pip install matplotlib
```

Note how the command started with an exclamation mark. This means it's a terminal command and not a Python-command. As an alternative you can use a percentage (which is quite different internally, but has the same effect).

You'll find this type of code block regularly on top of other notebooks. Usually they are commented out because they take up a lot of time when running the full notebook and should only be run once anyways.

The moment you have set up a complete virtual environment, it could be an idea to store this setup in a file. 


```python
!pip freeze > files/requirements.txt
```

When you now want to install all the selected libraries at once in a new venv, you import the setup file we just created and you can use the following code-block. (Note: here it's no use to run this command because all the libraries are already installed)

```python
!pip install -r files/requirements.txt
```
